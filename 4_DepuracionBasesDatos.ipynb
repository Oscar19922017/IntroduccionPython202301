El análisis exploratorio de datos (EDA, por sus siglas en inglés) es un proceso en el que se examinan y resumen los datos para obtener información útil y conocer mejor las características del conjunto de datos. El objetivo es descubrir patrones, tendencias, relaciones y valores atípicos en los datos.

El EDA implica una serie de pasos que se pueden resumir en los siguientes:

* Cargar y explorar los datos: El primer paso es cargar los datos y examinarlos para ver su estructura, tamaño, tipo y calidad.

* Limpiar los datos: Luego, se deben identificar y corregir los errores, valores faltantes o inconsistentes en los datos.

* Visualizar los datos: La visualización de los datos es una forma efectiva de descubrir patrones y tendencias en los datos. Las gráficas, histogramas, diagramas de dispersión, mapas de calor, entre otros, son herramientas que se pueden utilizar para explorar los datos.

* Analizar las relaciones: En este paso, se busca entender cómo se relacionan las variables del conjunto de datos. Se pueden utilizar técnicas como la correlación, regresión y análisis de varianza (ANOVA) para investigar estas relaciones.


* El EDA puede ayudar a identificar patrones y tendencias en los datos, así como posibles problemas o sesgos que puedan afectar la precisión de los modelos predictivos. También puede ayudar a los analistas a entender mejor el comportamiento de los clientes de crédito y a tomar decisiones informadas sobre la gestión del riesgo crediticio.

* En resumen, el análisis exploratorio de datos es una herramienta poderosa para descubrir información valiosa en los datos. Es una forma efectiva de comenzar a analizar datos y explorar patrones y relaciones sin tener que crear modelos predictivos complejos.

## **2.2 Diccionario De Variables**


Los datos "Default of Credit Card" proporcionan información sobre el comportamiento de pago de los clientes de tarjetas de crédito en Taiwán entre abril de 2005 y septiembre de 2005. El conjunto de datos incluye información sobre los clientes de crédito, como su edad, género, estado civil, nivel educativo, historial de pagos, límite de crédito y si incumplieron en su pago o no.

La variable objetivo es el incumplimiento del pago del cliente, que se define como la incapacidad de un cliente para pagar su deuda en un período de pago específico.

El conjunto de datos "Default of Credit Card" consta de 30.000 observaciones y 25 variables, y se ha utilizado en numerosos estudios de predicción de incumplimiento de pago.


* Id: identificación del cliente
* LIMIT_BAL: límite de crédito otorgado en dólares NT (nuevo taiwanés) (incluye el crédito individual y el familiar/suplementario)
* SEX: género (1 = masculino, 2 = femenino)
* EDUCATION: nivel educativo (1 = posgrado, 2 = universidad, 3 = escuela secundaria, 4 = otros)
* MARRIAGE: estado civil (1 = casado, 2 = soltero, 3 = otros)
* AGE: edad en años
* PAY_0 a PAY_6: historial de pagos anteriores de septiembre de 2005 a abril de 2005 (PAY_0 = el estado de reembolso en septiembre de 2005; PAY_2 = el estado de reembolso en agosto de 2005, y así sucesivamente hasta PAY_6 = el estado de reembolso en abril de 2005) (La escala de medición es la siguiente: -1 = pago debidamente, 1 = retraso en el pago por un mes, 2 = retraso en el pago por dos meses, y así sucesivamente hasta 8 = retraso en el pago por ocho meses, 9 = retraso en el pago por nueve meses y más)
* BILL_AMT1 a BILL_AMT6: Monto de la factura de septiembre de 2005 a abril de 2005 (BILL_AMT1 = cantidad de facturación en septiembre de 2005; BILL_AMT2 = cantidad de facturación en agosto de 2005, y así sucesivamente hasta BILL_AMT6 = cantidad de facturación en abril de 2005)
* PAY_AMT1 a PAY_AMT6: cantidad del pago anterior de septiembre de 2005 a abril de 2005 (PAY_AMT1 = cantidad pagada en septiembre de 2005; PAY_AMT2 = cantidad pagada en agosto de 2005, y así sucesivamente hasta PAY_AMT6 = cantidad pagada en abril de 2005)
* DEFAULT_PAYMENT_NEXT_MONTH: incumplimiento del pago (1 = sí, 0 = no)


fuente: https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset

## **B) Extructura de Control Bucles en Python**

En Python, existen dos tipos principales de bucles: for y while. Ambos permiten repetir una serie de instrucciones mientras se cumpla una determinada condición.

**Bucle for**

El bucle for se utiliza para recorrer una secuencia, como una lista, una tupla o un rango de números, tabla ect. Su sintaxis es la siguiente:

- for variable in secuencia:
- Cuerpo del bucle

```python
frutas=["manzana","pera","banana","piña","uva","durazno","papaya","granadilla","fresas"]
```

```python
len(frutas)
```

```python
frutas[0]
```

```python
frutas[1]
```

```python
frutas[2]
```

```python
for fruta in frutas:
    print(fruta)
```

En Python, es posible utilizar un bucle for con un contador utilizando la función range(). Esta función devuelve una secuencia de números en un rango especificado.

La sintaxis es la siguiente:

- for i in range(inicio, fin, paso):
-  Cuerpo del bucle

```python
for i in range(0,len(frutas),1):
    print(frutas[i])
```

## **Homologar SEX: género (1 = masculino, 2 = femenino)**

```python
def HomolarGenero(Genero):
    if (Genero==1):
        GeneroHomo="Hombre"
    elif (Genero==2):
        GeneroHomo="Mujer"
    else:
        GeneroHomo="Validar Informacion"
    return GeneroHomo
```

```python
HomolarGenero(3)
```

```python
GeneroHomologado=[]
for i in range(0,len(df),1): 
    GeneroHomologado.append(HomolarGenero(df["Genero"][i]))
```

```python
df["GeneroHomologado"]=GeneroHomologado
```
